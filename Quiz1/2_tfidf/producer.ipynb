```python
import pandas as pd
import re
from kafka import KafkaProducer
from kafka.errors import KafkaError
```

```python
f = open('book.txt', 'r', encoding='utf-8')
book = f.read()
pages = re.split(r'Page...\d+ Harry Potter and the Philosophers Stone - J.K. Rowling', book)
pages_clean = []

for i, page in enumerate(pages):
    page = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,”“—]", '', page).replace('\n', '').replace('’','\'').strip().lower()
    pages_clean.append((i,page))
```

```python
def write_to_kafka(topic_name, items):
  count=0
  producer = KafkaProducer(bootstrap_servers=['ec2-13-229-46-113.ap-southeast-1.compute.amazonaws.com:9092'])
  for key, message in items:
    try:
      producer.send(topic_name, key=str(key).encode('utf-8'), value=message.encode('utf-8'))
    except Exception as e:
      print(e)
    count+=1

    if count==2:
      break
  producer.flush()
  print("Wrote {0} messages into topic: {1}".format(count, topic_name))
```

```python
write_to_kafka("hope-pyspark-test", pages_clean)
```
